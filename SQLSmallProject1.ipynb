### Finding Densely Populated Countries

```python
%%sql
SELECT AVG(area) area_avg FROM facts;
```

### The table of countries with: 

* Population is above average.
* Area is below average.


```python
%%sql
SELECT * FROM facts
WHERE 
population > (select AVG(population) FROM facts)
AND 
area < (select AVG(area) FROM facts)
```

```python
%%sql
SELECT * from facts WHERE area_water/area < area_land/area 
```

### Top 5 countries with highest ratios of water to land

```python
%%sql
SELECT name country,
CAST(area_water as float)/area_land water_land_ratio
FROM facts
ORDER BY 2 DESC
LIMIT 5;
```

### Countries with more water than land

```python
%%sql
SELECT name country,area_water,area_land,CAST(area_water as float)/area_land water_land_ratio
FROM facts
WHERE area_water > area_land
```

### Top 5 Countries that will add the most people to their population next year

```python
%%sql
SELECT
name,
population_growth
FROM facts
ORDER BY 2 DESC LIMIT 5;
```

### Countries that have a higher death rate than birth rate

```python
%%sql
SELECT
name,
death_rate,
birth_rate,
death_rate - birth_rate delta
FROM facts
WHERE death_rate > birth_rate
ORDER BY delta DESC
```
